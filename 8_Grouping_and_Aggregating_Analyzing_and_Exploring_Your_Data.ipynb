# Grouping the Data

If we want to see specific results based on the country or based on some other column then we're going to have to group on that specific column and we have the group_by function for this.

Let's see the most popular social media sites for each country, in order to do this we're going to have to learn about grouping our data.

In the Pandas documentation it says that a group_by operation involves some combination of:



*   Splitting the object
*   Apply a function
*   Combine results





````python
# Just for a reference let's display the value counts for each country so that we can see the countries that have the most results for this particular survey.

# df['Country']

# This gives us the country that each respondent said that they were from and if we look at the value_counts for this then this is going to tally up all of the 
# unique responses:

df['Country'].value_counts()
````

Let's look at how to use the group_by function on our country column.

So first we are going to split the object and then we are going to apply a function and then it will combine those results.

````python
# Splitting the object.
# In this case we want to group all over results by country.
# The argument that we pass in always will be a list of columns (eve if it is just a column).

df.groupby(['Country'])
````

What we get back here is this DataFrameGroupBy object. So, what is this object and what exactly can we do with this?

This object contains a bunch of groups and to better understand what this is, let's take a look at an individual group that this DataFrame has:

````python
# Set the result as a variable 
country_grp = df.groupby(['Country'])
country_grp
````

````python
# Let's take a look at one of these groups.
# !!! Since we grouped our rows by country then we can grab a specific group by country name.

# Grab the group for the United States
country_grp.get_group('United States')
````

We get a DataFrame returned here with some survey results.

````python
# Look at the group for Mexico

country_grp.get_group('Mexico')
````

So that's what our DataFrameGroupBy object that we saw before consists of, it has broken up all of the different responses into groups by country name.

So this would be similar to running a filter on our original DataFrame:

````python
filt = df['Country'] == 'United States'
df.loc[filt]
````

## So if we're just looking to get information on a single country then it's very similar to just creating a filter like we did here, but instead of just grabbing the results for one country, group_by instead splits all of these responses by country name



Now that we have all of those split up and grouped by country name, now we can apply a function and bring those results back together.

So what kinf of function would we like to apply? Maybe we want to see the most popular social media sites broken down by country.


[Now if we just wanted to get the most popular social media sites by the United States or by India then we've already seen how we can do this. See code below]

````python
filt = df['Country'] == 'United States'
df.loc[filt, 'SocialMedia'].value_counts()
````

So these are the results for one specific country but if we were to run this on our DataFrameGroupBy object then it will give us the results for all of those country groups.

So if it helps you with how you think about this you can imagine that it's similar to running a filter and then applying a function like we did above with a single country, but when we group these using the group_by function and then apply a function then it will combine those groups to give us the results for all of those unique countries.

````python
country_grp['SocialMedia'].value_counts().head(50)
````

This is actually returning a Series and this Series has multiple indexes.

(The country is the first index)

````python
# Getting popular social media sites for Mexico
country_grp['SocialMedia'].value_counts().loc['Mexico']
````

````python
country_grp['SocialMedia'].count().loc['Mexico']
````

We cam see that those are the same results that we got before (like we were using filters), now you might be wondering well "hey if those are the same results that I got before then why is this even useful?"

And it's useful because now we can see this result with any country without running a filter on each individual country in the world, so for example if I wanted to see the most popular social media site for the United States then now instead of you know changing a filter over and over we can just look for the United States index.

````python
country_grp['SocialMedia'].value_counts().loc['United States']
````

````python
# With percentages
country_grp['SocialMedia'].value_counts(normalize=True).loc['Mexico']
````

Bringing back to what we were discussing at the beginning of the video we can also use this to run more traditional aggregate functions like mean, median and things like that so before we looked at the median salaries for the entire survey but now let's break these down by country instead.

# .agg() function

````python
# Maybe you're working on some analysis where you want to group your data but you also want to run multiple aggregate functions on your group.
# See the median and the mean of the salary column by country.

country_grp['ConvertedComp'].agg(['median', 'mean'])
````

````python
# See the median and the mean of the salary column for Mexico.
country_grp['ConvertedComp'].agg(['median', 'mean']).loc['Mexico']
````

Now depending on what you are trying to do you might run into some issues that you didn't quite expect so for example, let's say that you're trying to figure out how many people in each country know how to use Python, so before we do this to our group let's first look at how we do this with a single country using the filtering approach that we used earlier.

````python
# Se how many people know Python in a certain country.

filt = df['Country'] == 'Mexico'
df.loc[filt, 'LanguageWorkedWith'].str.contains('Python')
````

So this will return True for the rows that have Python in the LanguageWorkedWith column and false for the responses that don't.

This return a Series of True and False values where it tells us whether the LanguageWorkedWith column for each respondent contained contain that string of Python.

Now if we want to actually count the number of people who know Python then we can use the .sum() function to add all of these up.

Now normally you might think that some would only work with numerical data but some will also work on boolean, it will count all of the Trues as 1 and all of the False as 0, so to find out how many people know Python then we could simply just do a .sum():

````python
# Se how many people know Python in a certain country.

filt = df['Country'] == 'Mexico'
df.loc[filt, 'LanguageWorkedWith'].str.contains('Python').sum()
````

Now before when we wanted to run a similar aggregation function on our DataFrameGroupBy object we simply took the same approach on our GroupBy object, so for example, you might think that we could just do something like this to see how many people knew Python from each country:

````python
# We should able to do this:

country_grp['LanguageWorkedWith'].str.contains('Python').sum()
````

Previously we got a

```
Cannot  access attribute 'str' of 'SeriesGroupBy' objects, try using the 'apply' method.
```

But now we get a 

```
AttributeError: 'SeriesGroupBy' object has no attribute 'str'
```

So the reason that we get this error is because this is no longer a Series instead this is a SeriesGroupBy object and it tells us to instead it used the apply method.

When we run an apply method on a group object like this, we're going to specify a function that we want to be run on every series in this group.





````python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
````
