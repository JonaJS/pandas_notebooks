# TAB delimited file.

These are almost exactly the same thing as CSV files but instead of your data being separated by a comma, the data is instead separated by tabs so to do this in pandas we're still going to use the same CSV methods that we've already seen but we're going to pass in a custom separator **so we can write to a tab delimeted file** just by changing the file extension here:

```python
mexico_df.to_csv('modified.tsv', sep='\t') # we can also separate by -, but commas or tabs are probably the most common.
```

# SQL Databases

```python
# Connect to the DB
from sqlalchemy import create_engine # Work with DBs like sqlite | this will allow us to connect to the DB
import psycopg2 # Work with Postgre DB
```

Now that those are imported we should be able to create the engine which is basically our database connection.

```python
# Create engine
# postgre connection string
engine = create_engine('postgresql://dbuser:dbpass@localhost:5432/sample_db')
```

## Write or modify dataframe to a table in this DB.

And this table doesn't need to currently exist, by default it will create this table for us.

If it does already exists then we'll need to add abother argument to handle that.

```python
mexico_df.to_sql('table_name', engine)
```

```python
# Update the data and rewrite that data to the same table DB
# Write over a table.
mexico_df.to_sql('table_name', engine, if_exists='replace')
```

## Read

```python
sql_df = pd.read_sql('table_name', engine, index_col='Respondent')
```

## Reading using a SQL Query

```python
sql_df = pd.read_sql('SELECT * FROM table_name WHERE...', engine, index_col='Respondent')
```
